```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib 

df = pd.read_csv('/content/sample_data/result_csv.csv')

fig, ax = plt.subplots(figsize=(13,6))
ax2 = ax.twinx()

ax.set_title('Epochs vs. Loss and f1-Score')
ax.set_xlabel('No. of Epochs')
ax.plot(df['Epoch'], df['f1-Score'], color='green', marker='x')
ax2.plot(df['Epoch'], df['Loss'], color='red', marker='o')
ax.set_ylabel('f1-Score per epoch')
ax2.set_ylabel('Loss per epoch')
ax.legend(['f1-Score'],loc='upper right')
ax2.legend(['Loss'], loc='lower right')
matplotlib.rcParams.update({'font.size': 20})

ax.yaxis.grid(color='lightgray', linestyle='dashed')
plt.tight_layout()


# .pdf for your case, default .png
plt.savefig(r'/content/sample_data/fig1.pdf', bbox_inches='tight', dpi=300)
plt.show()
```
